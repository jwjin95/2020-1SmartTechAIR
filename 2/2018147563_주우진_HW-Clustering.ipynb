```python
import pandas as pd
import numpy
import xlrd
data = pd.read_excel('4_Cluster Analysis_Data.xlsx', sheet_name="Transactions")
```

```python
table = pd.pivot(data, index='Offer #', columns="Customer Last Name", values="Offer #")
```

```python
table = table.fillna(0)
table = table.astype(int)
table
```

```python
table[table>0]=1
table
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
kmeans3 = KMeans(n_clusters=3, max_iter=300).fit(table.T)
sil_3 = silhouette_score(table.T, kmeans3.labels_, metric='euclidean')

kmeans4 = KMeans(n_clusters=4, max_iter=300).fit(table.T)
sil_4 = silhouette_score(table.T, kmeans4.labels_, metric='euclidean')

kmeans5 = KMeans(n_clusters=5, max_iter=300).fit(table.T)
sil_5 = silhouette_score(table.T, kmeans5.labels_, metric='euclidean')
```

```python
sil_list = [sil_3,sil_4,sil_5]
print(sil_list)
print(sil_list.index(max(sil_list))+3)
```

```python
CList = [[0 for k in range(4)] for i in range(len(table.index))]
for i, vi in enumerate(table.index):
    for j, vj in enumerate(table.columns):
        CList[i][kmeans4.labels_[j]] += table.loc[vi][vj]
```

```python
cluster_table = pd.DataFrame(data=CList, index=table.index, columns=['Cluster 1','Cluster 2','Cluster 3','Cluster 4'])
```
