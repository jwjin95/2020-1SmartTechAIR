```python
import pandas as pd
import numpy as np
import xlrd
import math
book = xlrd.open_workbook('C:/Users/LG/Desktop/5_Naive Bayes_Data.xlsx')
sheet = book.sheet_by_index(0)
data = []
for i in range(sheet.nrows):
    data.append(sheet.row_values(i)[0])
```

```python
data = data[1:]
for i in range((len(data))):
    data[i] = data[i].lower()
    data[i] = data[i].replace(". "," ").replace(": "," ").replace("?"," ").replace("!"," ").replace(";"," ").replace(","," ")
```

```python
LK = []
for i in range(len(data)):
    LK.append(data[i].split(' '))
```

```python
LCK = []
for i in range(len(LK)):
    for j in range(len(LK[i])):
        LCK.append(LK[i][j])
```

```python
Lth3 = []
for i in range(len(LCK)):
    if len(LCK[i]) > 3:
        Lth3.append(LCK[i])
    else:
        pass
```

```python
Lth3
```

```python
LCKF = pd.DataFrame(Lth3,columns = ['token'])
c = LCKF.groupby(['token'])['token'].count()
c = pd.DataFrame(c)
c = c.rename(columns={'token':'count'})
c=c.reset_index()
c['add one']=c['count']+1
c['Prob']=c['add one']/c['add one'].sum()
c['LN(P)']=np.log(c['Prob'])
c
```

```python
LCKF = pd.DataFrame(Lth3,columns = ['token'])
d = LCKF.groupby(['token'])['token'].count()
d = pd.DataFrame(d)
d = d.rename(columns={'token':'count'})
d=d.reset_index()
d['add one']=d['count']+1
d['Prob']=d['add one']/d['add one'].sum()
d['LN(P)']=np.log(d['Prob'])
d
```

```python
sheet3 = book.sheet_by_index(2)
data3 = []
for i in range(sheet3.nrows):
    data3.append(sheet3.row_values(i)[2])
data3 = data3[1:]    
for i in range((len(data3))):
    data3[i] = data3[i].lower()
    data3[i] = data3[i].replace(". "," ").replace(": "," ").replace("?"," ").replace("!"," ").replace(";"," ").replace(","," ")
```

```python
LK = []
for i in range(len(data3)):
    LK.append(data3[i].split(' '))
```

```python
LK
```

```python
app_prob = [[] for cols in range(len(LK))]
for i in range(len(LK)):
    for j in range(len(LK[i])):
        if(len(LK[i][j])) > 3:
            if (c[c['token'].isin([LK[i][j]])]['LN(P)'].empty):
                app_prob[i].append(math.log(1/c['add one'].sum()))
            else:
                app_prob[i].append(c[c['token'].isin([LK[i][j]])]['LN(P)'].values[0])
        else:
            app_prob[i].append(0.0)
```

```python
app_prob
```

```python
other_prob = [[] for cols in range(len(LK))]
for i in range(len(LK)):
    for j in range(len(LK[i])):
        if(len(LK[i][j])) > 3:
            if (d[d['token'].isin([LK[i][j]])]['LN(P)'].empty):
                other_prob[i].append(math.log(1/d['add one'].sum()))
            else:
                other_prob[i].append(d[d['token'].isin([LK[i][j]])]['LN(P)'].values[0])
        else:
            other_prob[i].append(0.0)
```

```python
other_prob
```

```python
app_sum_prob = []
other_sum_prob = []
for i in range(len(app_prob)):
    app_sum_prob.append(sum(app_prob[i]))
    other_sum_prob.append(sum(other_prob[i]))
app_sum_prob
```

```python
other_sum_prob
```

```python
N=[]
for i in range(len(app_sum_prob)):
    if app_sum_prob[i] > other_sum_prob[i]:
        N.append('App')
    else:
        N.append('Other')
N
```
