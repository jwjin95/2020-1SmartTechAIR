```python
import pandas as pd
import numpy as np
import xlrd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_excel('champion_stats.xlsx',sheet_name='stats')
```

```python
pd.set_option('display.min_rows',200)
pd.set_option('display.max_rows',200)
```

```python
df
```

```python
df['stat_sum']=df['피해']+df['방어']+df['이동']+df['군중 제어']+df['보조']
df['피해']=df['피해']/df['stat_sum']
df['방어']=df['방어']/df['stat_sum']
df['군중 제어']=df['군중 제어']/df['stat_sum']
df['이동']=df['이동']/df['stat_sum']
df['보조']=df['보조']/df['stat_sum']
```

```python
df
```

```python
df.loc[df['피해 유형']=='물리','피해 유형']=0
df.loc[df['피해 유형']=='마법','피해 유형']=1
df.loc[df['피해 유형']=='혼합','피해 유형']=0.5
df
```

```python
pd.options.display.float_format = '{:.3f}'.format
df
```

```python
stats=df[['피해','방어','군중 제어','이동','보조']]
```

```python
kmeans3=KMeans(n_clusters=3, max_iter=300).fit(stats)
sil_3 = silhouette_score(stats, kmeans3.labels_, metric='euclidean')

kmeans4=KMeans(n_clusters=4, max_iter=300).fit(stats)
sil_4 = silhouette_score(stats, kmeans4.labels_, metric='euclidean')

kmeans5=KMeans(n_clusters=5, max_iter=300).fit(stats)
sil_5 = silhouette_score(stats, kmeans5.labels_, metric='euclidean')

kmeans6=KMeans(n_clusters=6, max_iter=300).fit(stats)
sil_6 = silhouette_score(stats, kmeans6.labels_, metric='euclidean')

print("sil_3: ",sil_3," sil_4: ",sil_4, " sil_5: ",sil_5, " sil_6: ",sil_6)
```

```python
df['cluster_by_4']=kmeans4.labels_
```

```python
df.sort_values(['cluster_by_4'])
```

```python
df.loc[df['피해 유형']== 0,'피해 유형']='물리'
df.loc[df['피해 유형']== 1,'피해 유형']='마법'
df.loc[df['피해 유형']== 0.5,'피해 유형']='혼합'
```
