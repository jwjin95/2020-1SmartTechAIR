```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_excel('champion_stats.xlsx',sheet_name='stats')
```

```python
df
```

```python
a=df[['피해','방어','군중 제어','이동','보조']]
df['sum']=df['피해']+df['군중 제어']+df['이동']+df['방어']+df['보조']
df
```

```python
kmeans3=KMeans(n_clusters=3).fit(a)
kmeans4=KMeans(n_clusters=4).fit(a)
kmeans5=KMeans(n_clusters=5).fit(a)
```

```python
df['cluster_by_3']=kmeans3.labels_
df['cluster_by_4']=kmeans4.labels_
df['cluster_by_5']=kmeans5.labels_
```

```python
pd.set_option('display.min_rows',200)
```

```python
pd.set_option('display.max_rows',200)
```

```python
df
```

```python
df.sort_values(['cluster_by_3'])
```

```python
sil_3 = silhouette_score(a, kmeans3.labels_, metric='euclidean')
sil_4 = silhouette_score(a, kmeans4.labels_, metric='euclidean')
sil_5 = silhouette_score(a, kmeans5.labels_, metric='euclidean')
print("sil_3: ",sil_3," sil_4: ",sil_4, " sil_5: ",sil_5)
```
